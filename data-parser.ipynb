```python
import pandas as pd
import numpy as np
from datetime import date
```

```python
mantas_df = pd.DataFrame(pd.read_csv('docs/data/counts.csv'))
mantas_df.head()
```

```python
mantas_df.rename(columns={"Group Size": "Group_Size"}, inplace=True)
```

```python
def get_lunar_day(full_moon_date, obs_date):
    month = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
    d0 = full_moon_date
    d1 = obs_date
    delta = d1 - d0
    degree = (delta.days/28.0) * 360.0
    return delta, degree
    
```

```python
lunar_days = []
lunar_deg = []
for i in range(mantas_df.shape[0]):
    temp = get_lunar_day(
    date(mantas_df['Full Yr'][i], mantas_df['Full Mo'][i],mantas_df['Full Da'][i]), 
    date(mantas_df['Yr'][i],mantas_df['Mo'][i], mantas_df['Da'][i]))
    lunar_days.append(temp[0])
    lunar_deg.append(temp[1])

mantas_df['lunar_day'] = lunar_days
mantas_df['lunar_deg'] = lunar_deg
mantas_df.head()
```

```python
def tide(pre_h, post_h, pre_t, post_t, time):
    tide = (time - pre_t)/(post_t-pre_t)*180   
    if((pre_h - post_h) < 0):  #rising
        tide += 180
    return tide
    
```

```python
tidal_phase = []
for i in range(mantas_df.shape[0]):
    temp = tide(mantas_df['Pre Height'][i], 
                mantas_df['Post Height'][i], 
                mantas_df['Pre Time'][i],
                mantas_df['Post Time'][i], 
                mantas_df['Time'][i])
    tidal_phase.append(temp)
    
mantas_df['Tidal_phase'] = tidal_phase
mantas_df.head()
```

```python
mantas_df.head()
```

```python
mantas_clean = mantas_df.drop(columns = ['Toby?', 'Full Yr', 'Full Mo', 'Full Da', 'Pre Time', 'Pre Height', 'Post Time', 'Post Height'])
mantas_clean.head()
```

```python
mantas_clean['ID'] = mantas_clean.index
mantas_clean.head()
```

```python
mantas_clean.to_csv('docs/data/clean_data.csv')
```

## Group by date

```python
polar_data = mantas_clean.drop(columns=["Latitude", "Longitude"])
print(polar_data.shape)
polar_data.tail()
```

```python
date_array = []
for i in range(polar_data.shape[0]):
    temp = date(polar_data['Yr'][i], polar_data['Mo'][i],polar_data['Da'][i])
    date_array.append(temp)

polar_data['date'] = date_array
polar_group = polar_data[['date', 'Group_Size']]
polar_group2 = polar_data[['date', 'lunar_day', 'lunar_deg', 'Tidal_phase']]
polar_group.tail()
```

```python
polar_group = polar_group.groupby('date').agg('sum')
print(polar_group.shape)

polar_group2 = polar_group2.groupby('date').agg('max')
print(polar_group2.shape)

if(polar_group2.shape[0] == polar_group.shape[0]):
    clean_polar = polar_group2
    clean_polar['Group_Size'] = polar_group['Group_Size']
else: 
    print("ERROR: did not agg correctly")
    
clean_polar.head()
```

```python
clean_polar.to_csv('docs/data/polar_data.csv')
```

## Group by lunar day for calendar

```python
calendar_data = clean_polar
calendar_clean = calendar_data.groupby('lunar_day').agg('mean')
calendar_clean.head()
```

```python
calendar_clean.to_csv('docs/data/calendar_data.csv')
```
